```python
arr=[]
with open("lc.txt","r")as f:
    for element in f:
        arr.append(element)
        
def remove_soln_tag(array,pattern):
    my_arr=[]
    for links in array:
        if pattern not in links:
            my_arr.append(links)
        else:
            print("removed"+links)
    return my_arr

arr=remove_soln_tag(arr,"/solution")
print(len(arr))
arr=list(set(arr))
with open('lc_problems2.txt', 'a') as f:
    # Iterate over each link in your final list
    for j in arr:
        # Write each link to the file, followed by a newline
        f.write(j)
```
