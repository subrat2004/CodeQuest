```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
s = Service('chromedriver.exe')
driver=webdriver.Chrome(service=s)
heading_class=".mr-2.text-label-1"
body_class=".px-5.pt-4"
index=1

def get_array_of_links():
    arr = []  # Array to store the lines of the file
# Open the file
    with open("lc_problems2.txt", "r") as file:
        # Read each line one by one
        for line in file:
            arr.append(line)
    return arr


def add_text_to_index_file(text):
    index_file_path = os.path.join("qdata2", "index.txt")
    with open(index_file_path, "a") as index_file:
        index_file.write(text + "\n")

        
def add_link_to_Question_index(texts):
    index_file_path=os.path.join("qdata2","qindex.txt")
    with open(index_file_path,"a", encoding="utf-8", errors="ignore")as file:
        file.write(texts)

def create_add_text_to_file(file_name,text):
    folder_path = os.path.join("qdata2", file_name)
    os.makedirs(folder_path, exist_ok=True)
    file_path = os.path.join(folder_path, file_name + ".txt")
    with open(file_path, "w", encoding="utf-8", errors="ignore") as new_file:
        new_file.write(text)

    


def getpage_data(url,index):
    #we have to use try except as some questions are premium
    try:
        driver.get(url)
        WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR,body_class)))
        time.sleep(1)
        heading=driver.find_element(By.CSS_SELECTOR,heading_class)
        body=driver.find_element(By.CSS_SELECTOR,body_class)
        print(heading.text)
        if(heading.text):
            add_text_to_index_file(heading.text)
            add_link_to_Question_index(url)
            create_add_text_to_file(str(index),body.text)
            return True 
    except Exception as e:
        print(e)
        return False
        
        
array=get_array_of_links()
for links in array:
    success=getpage_data(links,index)
    if(success):
        index+=1

driver.quit()  
```

```python

s = Service('chromedriver.exe')
driver=webdriver.Chrome(service=s)
body_class='.problem-statement'
index=1

#function to get all element links
def get_links_from_file():
    arr=[]
    with open('CF_cleaned.txt',"r")as f:
        for line in f:
            arr.append(line)
    return arr

#In case of codeforces as we are doing altogether
def add_links_to_index_file(texts):
    index_file_path=os.path.join("qdata_CF","qindex.txt")
    with open(index_file_path,"a", encoding="utf-8", errors="ignore")as file:
        file.write(texts)

        
def create_add_text_to_file(file_name,text):
    file_path = os.path.join('qdata_CF', file_name + ".txt")
    with open(file_path, "w", encoding="utf-8", errors="ignore") as new_file:
        new_file.write(text)

def get_text_from_links(url,index):
    #we have to use try except as some questions are premium
    try:
        driver.get(url)
        WebDriverWait(driver,10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR,body_class)))
        time.sleep(1)
        body=driver.find_element(By.CSS_SELECTOR,body_class)
        add_links_to_index_file(url)
        create_add_text_to_file(str(index),body.text)
        return True 
    except Exception as e:
        print(e)
        return False
        
array=get_links_from_file()
for link in array:
    success=get_text_from_links(link,index)
    if(success):
        index+=1
    
    
driver.quit()   
       
    
```
